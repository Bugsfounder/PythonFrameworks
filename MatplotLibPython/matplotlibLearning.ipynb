## Plot Scatter Generating Random Data Using Some numpy Functions and others

```python
np.random.seed(19680801) # seed the random number generator
data = {
    'a':np.arange(50),
    'c':np.random.randint(0,50,50),
    'd': np.random.randn(50)
}
data['b'] = data['a']+10*np.random.randn(50)
data['d'] = np.abs(data['d'])*100

fig, ax = plt.subplots(figsize=(5,2.7), layout="constrained")
ax.scatter('a', 'b', c='c', s='d', data=data)
ax.set_xlabel("Entry A")
ax.set_ylabel("Entry B")
```

## Plot Graph 

```python
x = np.linspace(0,2,100)
# Note that even in the 00-style, we use `.pyplot.figure` to create the Figure.
fig, ax = plt.subplots(figsize=(5,2.7), layout="constrained")
ax.plot(x, x, label="Linear") # Plot some data on the axes.
ax.plot(x, x**2, label="Quadratic") # Plot more data ont he axes...
ax.plot(x, x**3, label="Cubic") # ...  and some more.
ax.set_xlabel('X Label') # Add an x-label to the axes.
ax.set_ylabel('Y Label') # add a y-label to the axes.
ax.set_title("Simple Plot") # Add a title to the axes.
ax.legend()
```

## Helper Function

```python
def my_plotter(ax, data1, data2, param_dict):
    """
     A helper function to make a graph.
    """
    out = ax.plot(data1, data2, **param_dict)
    return out
```

```python
data1, data2, data3, data4 = np.random.randn(4, 100) # make 4 random data sets with 100 entries
```

```python
fig, (ax1, ax2) = plt.subplots(1,2, figsize=(12, 4))
my_plotter(ax1, data1, data2, {'marker': 'x'})
my_plotter(ax2, data3, data4, {'marker': 'o'})
```

## my graph

```python
data1, data2, data3, data4, data5, data6 = np.random.randn(6, 100)
fig, (axes1, axes2, axes3) = plt.subplots(1,3, figsize=(30,8))

# plotting first graph ----> 1
axes1.plot(data1, data2, marker='x', label="1")
axes1.set_title("First Graph")
axes1.set_xlabel("X Entry")
axes1.set_ylabel("Y Entry")
axes1.legend()

# plotting second graph ----> 2
axes2.plot(data3, data4, marker="o", label="2")
axes2.set_title("Second Graph")
axes2.set_xlabel("X Entry")
axes2.set_ylabel("Y Entry")
axes2.legend()

# plotting third graph ----> 3
# more on markers: https://matplotlib.org/3.1.1/api/markers_api.html
# axes3.plot(data5, data6, marker="s") 
axes3.plot(data5, data6, marker="P", label="3")
axes3.set_title("Third Graph")
axes3.set_xlabel("X Entry")
axes3.set_ylabel("Y Entry")
axes3.legend()
```

```python
fig, a = plt.subplots(figsize=(12,6))
a.plot(data1, data2, marker='s', label="Manisha Graph" )
a.plot(data3, data4, marker='s', label="Manisha Graph 2" )
a.legend()
```
